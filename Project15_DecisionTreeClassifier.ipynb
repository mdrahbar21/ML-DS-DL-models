```python
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn import tree


df = sns.load_dataset('iris')
X = df.iloc[:,0:4].values
Y = df.iloc[:,4]

cf=DecisionTreeClassifier()
cf.fit(X,Y)

decPlot=plot_tree(decision_tree=cf, feature_names=["sepal_length","sepal_width","petal_length","petal_width"],class_names=["setosa","versicolor","verginica"], filled=True, precision=4, rounded=True)

text_representation=tree.export_text(cf,feature_names=["sepal_length","sepal_width","petal_length","petal_width"])
print(text_representation)
```
