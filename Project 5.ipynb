```python
import numpy as np
from sklearn import datasets
import pandas as pd
import seaborn as sns
```

```python
bostonset=datasets.load_boston()
# we are loading the boston dataset

boston=pd.DataFrame(bostonset.data, columns=bostonset.feature_names)
# we are making a 2 dimensional data with columns as boston features and rows as data

corr=boston.corr().round(2) 
# .round(2) is rounding off upto 2 decimal places
# also, here we are setting a correlation between boston features
```

```python
sns.heatmap(data=np.abs(corr),annot=True, fmt="1.1f")
# we are visualising the **absolute** data of correlation matrix with fmt (format) as upto 1 decimal place
```

```python
sns.heatmap(data=np.abs(corr),annot=False, fmt="1.1f")
```
