```python
from sklearn import datasets
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.metrics import confusion_matrix
```

```python
irisset= datasets.load_iris()
X=irisset.data
Y= irisset.target
```

```python
pca=PCA(n_components=2)
Xp=pca.fit(X).transform(X)
```

```python
plt.figure(1)
plt.scatter(Xp[:, 0], Xp[:,1], c=Y, cmap='jet',s=10)
plt.suptitle('Original Clusters')
plt.grid(1, which='both')
plt.axis('tight')
plt.show()
```

This code is performing principal component analysis (PCA) on the iris dataset and then visualizing the result using matplotlib

Here is a brief breakdown of the code:

`datasets.load_iris()` loads the iris dataset from scikit-learn's datasets module <br>
<br>
`irisset.data` extracts the feature data from the iris dataset, which contains measurements of sepal length, sepal width, petal length, and petal width for each sample.
<br><br>
`irisset.target` extracts the target data from the iris dataset, which contains the corresponding species labels for each sample (setosa, versicolor, or virginica).
<br><br>
`PCA(n_components=2)` creates an instance of the PCA class with 2 components, which will be used to reduce the dimensionality of the feature data from 4 to 2
<br><br>
`pca.fit(X)` fits the PCA model to the feature data.
<br><br>
`pca.transform(X)` applies the trained PCA model to the feature data to obtain the transformed data with reduced dimensionality.
<br><br>
`plt.scatter(Xp[:, 0], Xp[:,1], c=Y, cmap='jet',s=10)` creates a scatter plot of the transformed data, where the x and y coordinates correspond to the first and second principal components, respectively. The color of each point is determined by its species label (setosa, versicolor, or virginica), and the colormap is set to 'jet'. The **size** of each point is set to 10.
<br><br>
`plt.suptitle('Original Clusters')` adds a main title to the plot.
<br><br>
`plt.grid(1, which='both')` adds a grid to the plot.
<br><br>
`plt.axis('tight')` adjusts the axis limits to fit the data.
<br><br>
`plt.show()` displays the plot.
<br><br>
Overall, this code is a simple example of how to perform PCA on a dataset and visualize the result using matplotlib. The plot shows how the different species of iris samples are distributed in the reduced 2D space, which can provide insights into the structure of the data and help with further analysis.
<br><br><br><br>
The `fit()` method of the PCA class learns the principal components of the data by calculating the **eigenvectors and eigenvalues of the covariance matrix of the input data X**. These eigenvectors are sorted in descending order of the corresponding eigenvalues, such that the **first principal component** captures the **maximum amount of variance in the data**, the **second principal component captures the next maximum amount of variance**, and so on.
<br><br>
After fitting the PCA model, the eigenvectors and eigenvalues are stored as attributes of the PCA object. These can be accessed using pca.components_ and pca.explained_variance_, respectively.
<br><br>
The `transform()` method of the PCA class then applies the trained PCA model to the input data X to obtain the transformed data Xp. This is done by **projecting the input data onto the learned principal components**
<br><br>
In this case, we have specified `n_components=2`, which means that we want **to reduce the dimensionality of the data from 4 to 2**. Therefore, the **`transform()`** method **returns the transformed data with only 2 columns**, corresponding to the first and second principal components, respectively.
<br><br>
Overall, the fit() and transform() methods of the PCA class are used together to perform dimensionality reduction on the input data, while preserving as much of the original information as possible.
<br><br><br>


<h3> target() </h3>

<br>

`irisset.target` is a attribute of the iris dataset object irisset that contains the target variable, which is the species of the iris flowers that the samples belong to
<br><br>
In the iris dataset, there are three classes of iris flowers: Setosa, Versicolor, and Virginica. For each sample in the dataset, the target variable specifies which class the sample belongs to. The target variable is represented as an integer value for each sample, where:
<br><br>
**0 represents Setosa** 
<br><br>
**1 represents Versicolor** 
<br><br>
**2 represents Virginica** 
<br><br>
The irisset.target attribute returns an array of target values, with one value for each sample in the dataset. In the code snippet, Y = irisset.target assigns the target values to the variable Y. This variable is later used to specify the color of each point in the scatter plot based on its corresponding target value
<br><br>

Overall, irisset.target is an important attribute of the iris dataset that provides information about the class labels for each sample in the dataset. It is used in many machine learning tasks, such as classification, where the goal is to predict the target variable (i.e., the species of an iris flower) based on the input features (i.e., sepal length, sepal width, petal length, and petal width).

```python
GMM=GaussianMixture(n_components=3)
GMM.fit(Xp)
y_predG=GMM.predict(Xp)
```

```python
plt.figure(2)
plt.scatter(Xp[:, 0], Xp[:,1], c=y_predG, cmap='jet',s=10)
plt.suptitle('GMM Clusters')
plt.grid(1, which='both')
plt.axis('tight')
plt.show()
```

```python
cmat=confusion_matrix(Y,y_predG)
cmat
```
