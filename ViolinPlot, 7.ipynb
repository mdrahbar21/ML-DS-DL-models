```python
import pandas as pd
import seaborn as sns
from sklearn import datasets
tips = sns.load_dataset("tips")
```

```python
sns.violinplot( x="total_bill",hue="smoker", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill",hue="smoker", data=tips)
```

```python
sns.violinplot(x="day", y="total_bill",hue="smoker", data=tips, split=True)
```

```python
sns.violinplot(x="day", y="total_bill",hue="smoker", data=tips, split=True, inner="quartile")
```

```python
sns.violinplot(x="day", y="total_bill",hue="smoker", data=tips, split=True, inner="stick")
```

```python
sns.swarmplot(x="total_bill",data=tips)
```

```python
sns.swarmplot(x="total_bill",data=tips,y="day")
```

```python
sns.swarmplot(x="total_bill",data=tips,y="day", hue="sex")
```

```python
ax=sns.boxplot(x="total_bill", y="day", data=tips)
ax=sns.swarmplot(x="total_bill", y="day", data=tips, color='.2')
```

```python
ax=sns.violinplot(x="total_bill", y="day", data=tips)
ax=sns.swarmplot(x="total_bill", y="day", data=tips, color='.2')
```
