```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn import datasets

irisset = datasets.load_iris()
X=irisset.data[:50,:2]

reg=LinearRegression().fit(X[:,0:1],X[:,1])
w=reg.coef_
c=reg.intercept_

xpoints=np.linspace(4,6)
ypoints=w[0]*xpoints + c;
plt.plot(xpoints, ypoints, 'g-')
plt.scatter(X[:, 0], X[:, 1],s=10)
plt.suptitle('Linear Regression IRIS Data')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.grid(1,which='both')
plt.axis('tight')
plt.show()
```

**Here, we are finding the best fit linear relation between 2 features using linear regression**

The necessary libraries are imported:
<br><br>
The Iris dataset is loaded using the load_iris() function from the datasets module. The dataset contains information about different Iris flowers, including their sepal length and width.
<br><br>
The first 50 samples of the Iris dataset are selected and only the first two features (sepal length and sepal width) are extracted. These samples are stored in the variable X.
<br><br>
An instance of the LinearRegression class is created, named reg.
<br><br>
The fit() method of the LinearRegression class is called with the training data X[:,0:1] (sepal length) as the independent variable and X[:,1] (sepal width) as the dependent variable. This method fits a linear regression model to the data.
<br><br>
The coefficients of the linear regression model are stored in the variable w, and the intercept is stored in the variable c.
<br><br>
An array of evenly spaced values between 4 and 6 is generated using linspace() from numpy. These values represent the x-coordinates for plotting the regression line.
<br><br>
The corresponding y-coordinates for the regression line are computed using the equation of a straight line: ypoints = w[0] * xpoints + c. These values are stored in the variable ypoints.
<br><br>
The plot() function from pyplot is used to plot the regression line (xpoints vs ypoints) as a green line.
<br><br>
The scatter() function from pyplot is used to create a scatter plot of the original data points, where X[:, 0] represents the x-coordinates (sepal length) and X[:, 1] represents the y-coordinates (sepal width).
<br><br>
Various formatting commands are used to add labels, title, and gridlines to the plot.
<br><br>
Finally, the plot is displayed using the show() function from pyplot.
<br><br>
In summary, this code loads the Iris dataset, performs linear regression on the sepal length and width data, and plots the regression line along with the original data points.
